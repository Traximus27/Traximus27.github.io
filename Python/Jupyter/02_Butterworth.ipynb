```python
import numpy as np
import control 
from matplotlib import pyplot as plt 
```

```python
s = control.tf('s')
Part1 = 1/(s**2 + 0.518*s +1)
#Part1 = control.tf([1],[1, 0.518, 1])
Part2 = 1/(s**2 + 1.414*s +1)
Part3 = 1/(s**2 + 1.932*s +1)
Tot = (Part1*Part2*Part3);

w = np.arange(0.1,10,0.01)

OUT = control.bode_plot(Part1, w , dB=  True, plot = True)
OUT = control.bode_plot(Part2, w , dB=  True, plot = True)
OUT = control.bode_plot(Part3,w , dB=  True, plot = True)
#OUT = control.bode_plot(Tot, w , dB=  True, plot = True)
```

```python
OUT = control.bode_plot(Tot,w , dB=  True, plot = True)
control.pzmap(Tot, grid = True) 
```

```python
TotNew2 = Part1*PartNew2
OUT = control.bode_plot(Part1, w , dB=  True, plot = True)
OUT = control.bode_plot(TotNew2, w , dB=  True, plot = True,)
```

```python
OUT = control.pzmap(TotNew2, grid = True)
```
