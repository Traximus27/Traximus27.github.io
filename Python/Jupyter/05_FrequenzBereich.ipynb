```python
import control 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a0 = 1
a1 = 10
a2 = 31
a3 = 30
b0= 1
```

```python
num = np.array(b0)
den = np.array([a3, a2, a1, a0])
strecke = control.tf(num,den)
regler = control.tf([5, 1],[5, 0])
G0 = regler*strecke
```

```python
control.bode_plot(G0)
Gw = control.feedback(G0,1);
control.bode_plot(Gw)
```

```python
t,y = control.step_response(Gw)
control.step_info(Gw)
plt.plot(t,y)
plt.grid()
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```
