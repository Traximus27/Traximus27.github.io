```python
import numpy as np
from matplotlib import pyplot as plt
import control 


# Local Program 
def mySystem1(Tau,t):
    sys = control.tf(1,[Tau, 1])
    t,y = control.step_response(sys,t)
    return y 

def mySystem2(eps,wn,t):
    sys = control.tf(wn*wn,[1, 2*wn*eps, wn*wn])
    t,y = control.step_response(sys,t)
    return y 

def mySystem(Type, List,t):    
    if Type == 1:
        Tau = List[0]
        y = mySystem1(Tau,t)
    else:
        eps = List[0]
        wn =  List[1]
        y = mySystem2(eps,wn,t)
    return y 

# Main Program 
Tau = 5
wn = 0.5
eps = 0.57
t = np.arange(0,25,0.1)
List = [eps, wn]
y = mySystem(2,List,t)
```

```python
plt.figure() 
plt.plot(t,y,label = 'My System')
plt.xlabel(' Time [s]')
plt.ylabel('Output y')
plt.legend()
plt.grid()
plt.show()
```
