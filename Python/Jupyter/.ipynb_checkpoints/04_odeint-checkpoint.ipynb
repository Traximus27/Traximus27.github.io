```python
from scipy.integrate import odeint 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def deriv(xa, t, xe): 
    a0 = 0.7;
    b0 = 0.2;
    xapunkt = -a0*xa + b0*xe
    return xapunkt
```

```python
def deriv2(x,time,u):
    Tau = 5 
    K = 2
    xdot = -(1/Tau)*x + (K/Tau)*u
    return xdot
```

```python
time = np.arange(0, 10, 0.01)
xaAnfang = 0
u = 1; 
xa = odeint(deriv, xaAnfang, time, args=(u,))
```

```python
time = np.arange(0,25,0.001)
x0 = 2
u = 10
yp = odeint(deriv2,x0,time, args=(u,))
```

```python
plt.figure()
plt.plot(time, yp)
plt.grid()
plt.show()
```
